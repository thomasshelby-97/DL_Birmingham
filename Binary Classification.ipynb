## **Binary Classification using Multi Layer Perceptron**

```python
# Importing the necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
```

### **Accessing the Dataset**

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 400, n_features = 2, centers = 2, random_state = 100)
```

```python
# Displaying the feature set

X
```

```python
# Displaying the target

y
```

### **Visualization of the Data**

```python
plt.scatter(X[:, 0], X[:, 1], c = y); # Separating the dataset using the color of the target variable
```

### **Splitting the data into train and test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
# Displaying the shape of the train and test data

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### **Building the Neural Network Model**

```python
model_1 = tf.keras.Sequential() # Sequential is a class within Keras.
# keras has been bundled on top of tensor flow to make it more effective. Keras is a subset of tensorflow. Earlier both were different.
# We will be focussing on the Sequential architecture of Neural Network. We try to create an object of Sequential class.
```

```python
# Adding the hidden layer

model_1.add(tf.keras.layers.Dense(8, activation = 'relu')) 
# 8 is the no of units in the hidden layer. We can choose the no of units based on our requirement.
# Linear activation function / None is the default activation function if nothing is mentioned.
# We want a dense layer architecture. Dense means closely connected, i.e., the node is connected to each and every unit of the next layer.
```

```python
# Adding the output layer

model_1.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
# In our case, we have a single output which is connected with the 8 units of the hidden layers.
# We choose sigmoid to be the activation function which gives the output between 0 and 1. Based on the threshold (0.5 as default), the data is classified.
```

### **Compiling the model**

```python
model_1.compile(optimizer = tf.keras.optimizers.Adam(),
                loss = tf.keras.losses.BinaryCrossentropy(),
                metrics = ['accuracy'])
# When we say compile, we need to state the loss function. We are going to use Binary Cross Entropy loss function.
# Adding certain characteristics (optimizers), the loss function will be optimized. The example of optimizers: rmsprop, ols, Adam etc.
# Compiling of a model also takes into the consideration the performance metrics which the user defines.
```

### **Training the model**

```python
tf.random.set_seed(100)
history = model_1.fit(X_train, y_train, epochs = 100)
hist = pd.DataFrame(history.history)
hist.plot();
# epochs refer to the number of iterations and interaction between layers in the forward and backward direction.
# We use history to store the train data. This is done to compare the results when the epochs is changed in the subsequent iterations.
# During each epoch, the weight is ascertained between the layers. In this way the process continues till the epochs value which is provided.
# This is a time consuming process and in order to reduce the time, we go for batch sizing.
# By default, the batch size is taken to be 32. Hence, when we have 320 data points, the epoch value shows (320 / 32 = 10).
```

### **Model Summary**

```python
model_1.summary()
# By default the machine assigns one bias, hence in the first layer we have (2 + 1) * 8 = 24 parameters.
# While in the next layer we have (8 + 1) * 1 = 9 parameters.
```

### **Plotting the model**

```python
from tensorflow.keras.utils import plot_model

plot_model(model_1, show_shapes = True)
```

### **Prediction of the model**

```python
model_1.predict(X_test)
```
